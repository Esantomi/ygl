#### 결측치 처리는 행 삭제, 결측치 대체로 가능하다. (대체의 경우 평균, 중앙값이 쉽다.)

```python
m = tt['age'].median()  # 중앙값 m에 할당
```

```python
tt['age'] = tt['age'].fillna(m)  # NaNdmf m으로 채우기
```

```python
tt.describe()  # age 714개가 891개로 꽉 찼다.
```
