# seaborn으로 데이터 시각화하기 2 : tips

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## pairplot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'size')  # 몇 명이 왔는지로 구분
```

```python
sns.pairplot(tips, hue = 'size', height = 4)  # plot 크기
```

```python
sns.pairplot(tips, hue = 'day')
```

```python
sns.pairplot(tips, hue = 'day', height = 3)
```

## barplot

```python
sns.barplot(x = 'day', y = 'tip', data = tips)
```

```python
# 요일별로 정보 카운트
tips['day'].value_counts()
```

```python
# 특정 feature를 색깔로 구분
sns.barplot(x = 'day', y = 'tip', hue = 'smoker', data = tips)
```

## lmplot
- 선형 관계 살필 때 사용

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# 흡연자에 대해 구분해서 보기
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)
```

```python
# size별로 그래프를 보이기
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size', data = tips)
```

```python
# 한 줄당 플롯 수
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size', col_wrap = 3, data = tips)
```

## pivot
- groupby와 유사하다.

```python
pivot = tips.pivot_table(index = 'day', columns = 'smoker', values = 'tip')
```

```python
pivot
```

```python
# heatmap으로 그려 보기
sns.heatmap(pivot)
```

```python
sns.heatmap(pivot, annot = True)
```

```python
sns.heatmap(pivot, annot = True, cmap = 'Blues')
```

## scatterplot

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# 크기로 size(몇 명) 구분
sns.scatterplot(x = 'total_bill', y = 'tip', size = 'size', data = tips)
```

```python
# 흡연 유무로 구분
sns.scatterplot(x = 'total_bill', y = 'tip', size = 'smoker', data = tips)
```

## relplot

```python
# 산점도
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# 요일 별로 색깔 구분
sns.relplot(x = 'total_bill', y = 'tip', hue = 'day', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'day', col = 'size', data = tips)
```

## catplot
- categorical data에 많이 쓰는 plot
- 주로 x축은 categorical, y축은 수치 데이터

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# 흡연 유무에 대해 total_bill
sns.catplot(x = 'smoker', y = 'total_bill', data = tips)
```

```python
# 흡연 유무에 대해 tip
sns.catplot(x = 'smoker', y = 'tip', data = tips)
```
