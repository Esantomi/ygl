```python
# Selenium
from selenium import webdriver

driver = webdriver.Chrome("./crawling_practice/chromedriver")
# ipynb 파일과 chromedriver.exe 파일이 같은 폴더에 있으면 상대 경로(./)로 경로 설정 가능
driver.get("https://datalab.naver.com/")


# BeautifulSoup
from bs4 import BeautifulSoup

# driver.page_source # 홈페이지 내용
html = driver.page_source
html_parsing = BeautifulSoup(html, "lxml")
# print(html_parsing)


# 일자
title = html_parsing.find("span", {"class":"title_cell"})
print("일자 : " + title.text)

# 검색어 순위
ranking = html_parsing.find_all("em", {"class":"num"})
rank_list = []

for rank in ranking:
    rank_list.append(rank.get_text())
    
keywords = html_parsing.find_all("span", {"class":"title"})
keyword_list = []

for keyword in keywords:
    keyword_list.append(keyword.get_text())

for i in zip(rank_list, keyword_list):
    print(i)


# 드라이버 종료
driver.close()
```
