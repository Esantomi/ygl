```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

### groupby()

#### 종별 평균

```python
# 종별로 묶어서(group) 통계 정보 계산 및 그래프 그리기

iris.groupby(iris['species']).mean()  # 종별 평균 통계 정보
```

```python
df = iris.groupby(iris['species']).mean()
```

```python
# 선 그래프
df.plot()
```

```python
# 막대 그래프
df.plot.bar()
```

#### 종별 중앙값

```python
# 종별 묶어 중앙값 보기

df2 = iris.groupby(iris['species']).median()
```

```python
df2
```

```python
df2.plot()
```

```python
df2.plot.bar()
```

```python
# x label 옆으로 돌려 주기

df2.plot.bar(rot = 0)  # rotate
```

### 산점도

```python
iris.plot.scatter(x = 'sepal_length', y = 'sepal_width')
```

```python
# 조건 where
x = np.where(iris['species'] == 'setosa', 'red', 'green')
```

```python
iris.plot.scatter(x = 'sepal_length', y = 'sepal_width', c = x)
```

```python
# 조건 추가
x = np.where(iris['species'] == 'setosa', 'red',
    np.where(iris['species'] == 'versicolor', 'green', 'blue'))
```

```python
iris.plot.scatter(x = 'sepal_length', y = 'sepal_width', c = x)
```

```python
# 꽃잎이 좀 더 선형 관계, 꽃잎으로 그려 보자.

iris.plot.scatter(x = 'petal_length', y = 'petal_width', c = x)
```
